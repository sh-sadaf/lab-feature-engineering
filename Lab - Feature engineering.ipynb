```python
data = data.apply(lambda x: x.replace(" ", np.NaN)) # in order to replace all empty values with nan. 
```

```python
# replacing missing values with mode
data['GEOCODE2'].fillna(data['GEOCODE2'].mode()[0],inplace = True)
data['WEALTH1'].fillna(data['WEALTH1'].mode()[0], inplace=True)
data['ADI'].fillna(data['ADI'].mode()[0], inplace=True)
data['MSA'].fillna(data['MSA'].mode()[0], inplace=True)
data['DMA'].fillna(data['DMA'].mode()[0], inplace=True)
```
